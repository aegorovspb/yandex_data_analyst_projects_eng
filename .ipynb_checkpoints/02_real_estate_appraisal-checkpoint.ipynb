## Data preprocessing

1. Rename the column 'citycenter_nearest'
2. Fill in the empty values of 'balcony' zero values. We will proceed from the assumption that the missing data simply did not indicate the number because there was no balcony.
3. Fill in the empty values of 'locality_name'. Let's use the group by method to group values and search for the most popular value, since there are only 49 (0.2%) missing rows, then these data can be neglected and filled in for the convenience of further work.

```python
# rename 'cityCenters_nearest'
df = df.rename(columns={'cityCenters_nearest': 'city_сenter_nearest'})

# fill in the values of 'balcony' zero values
df['balcony'] = df['balcony'].fillna(value=0)

# find the most popular value for 'locality_name'
data_grouped = df.groupby('locality_name')['locality_name'].count().sort_values(ascending=False)
display(data_grouped)

# fill in the empty values of 'locality_name' with the most popular value of St. Petersburg
df['locality_name'] = df['locality_name'].fillna('Санкт-Петербург')
```

9. Fill in the empty values: floors_total, kitchen_area, airport_nearest, city_center_nearest, parks_around 3000, parks_nearest, ponds_around3000, ponds_nearest with median values. Based on the fact that the number of passes is about 20%, we will not devote too much time to processing these passes. If it turns out that these characteristics are of key importance, it will be possible to conduct additional research with a more detailed processing of these data.

```python
columns_to_replace_median = ['floors_total', 'kitchen_area', 'airports_nearest',
                             'city_сenter_nearest', 'parks_around3000', 'parks_nearest',
                             'ponds_around3000', 'ponds_nearest']

count_column_median = 0            # number of measurements
wrong_column_median = 0            # number of invalid values
wrong_column_content_median = []   # empty list for invalid values 

for column in columns_to_replace_median:
    try:
        # iterating over column names in the median calculation cycle
        median_index = df[column].median()
        df[column] = df[column].fillna(value=median_index)
        count_column_median += 1
    except:
        wrong_column_content_median.append(column)
        wrong_column_median +=1 

print('Number of changes', count_column_median )
print('Invalid lines', wrong_column_content_median)
print('Number of invalid rows', wrong_column_median)
if  wrong_column_median == 0:
    print('The specified data has been successfully replaced with median data!')
```

```python
# checking all the changes made
df.info()
```

10. From the above data it can be seen that 'floors_total' (all floors), 'balcony' (balcony), 'days_exposition'(duration of sale), 'ponds_nearest' (ponds nearby), 'parks_nearest' (parks nearby)

Should have only integer values, replace their type with an integer.

```python
# replace the real data type with an integer one
# use an algorithm to handle possible errors

columns_to_replace_int = ['floors_total', 'balcony', 'days_exposition', 'ponds_nearest', 'parks_nearest']

count_column_int = 0            # count of measurements
wrong_column_int = 0            # count of errors
wrong_column_content_int = []   # list for invalid values

for column in columns_to_replace_int:
    try:
        # iterating over column names in a loop and replacing the type with int
        df[column] = df[column].astype('int')
        count_column_int += 1
    except:
        wrong_column_content_int.append(column)
        wrong_column_int +=1 

print('Number of changes', count_column_int )
print('Invalid lines', wrong_column_content_int)
print('Number of invalid rows', wrong_column_int)
if  wrong_column_int == 0:
    print('The specified data has been successfully replaced with integer data!')
```

```python
# checking the changes made
df.info()
```

```python
# let's check the data for negative values
columns_to_zero = ['total_images', 'last_price', 'total_area',
       'rooms', 'ceiling_height', 'floors_total', 'living_area', 'floor',
       'kitchen_area', 'balcony',
       'airports_nearest', 'city_сenter_nearest',
       'parks_around3000', 'parks_nearest', 'ponds_around3000',
       'ponds_nearest', 'days_exposition']

count_column_zero = 0            # number of measurements
wrong_column_zero = 0            # number of values less than zero
wrong_column_content_zero = []   # empty list for invalid values 

for column in columns_to_zero:
    try:
        # checking negative values
        columns_to_zero = df[df[column] < 0]
        count_column_zero += 1
    except:
        wrong_column_content_zero.append(column)
        wrong_column_zero +=1 
print('Number of measurements', count_column_zero )
print('Invalid lines', wrong_column_content_zero)
print('Number of invalid rows', wrong_column_zero)
if  wrong_column_zero == 0:
    print('No negative values were found!')
```

## Calculations and adding results to the table.

```python
# Calculate and add to the table the price per square meter - last_price_square_meter is the ratio of two quantities
# To do this, use the values from last_price (price at the time of sale) and total_area (total area)
df['last_price_square_meter'] = df['last_price'] / df['total_area']
display(df['last_price_square_meter'].head())
```

```python
# Add new data: day of the week, month and year of publication of the ad;
# Let's use the to_datetime() method
# Names of new columns weekday - day of the week, month - month, year - year

df['weekday'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d %H:%M').dt.weekday
print(df['weekday'].head(5))

df['month'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d %H:%M').dt.month
print(df['month'].head(5))

df['year'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%d %H:%M').dt.year
print(df['year'].head(5))
```

```python
# create a new floor_category column
# add new apartment floor data to it; options — first, last, other;
# to do this, use the floor_group function
# We will go through all the floors of the apartment, while 
# ground floor - 'min' is the minimum value
# the last floor is a value equal to the value in floors_total (total floors)

# To start, let's find the minimum and maximum value
floor_min = df['floor'].min()
floor_max = df['floor'].max()
print('Minimum floor', floor_min)
print('Maximum floor', floor_max)

# let's create a function, floor_group, which will take a new value based on incoming data
# new value - group, incoming data - total floors
# 'floors_total' == the last floor for this house
def floor_group(income_index):
    
    if income_index == floor_min:
        return 'first'
    if income_index == df.loc[income_index, 'floors_total']:
        return 'last'

    return 'another' 

# print(df.loc[test, 'floors_total']) - accessing a single cell
# create a new column in which we will put the result of the function
# function we go through the values of the 'floor' column
df['floor_category'] = df['floor'].apply(floor_group)

# to check, we output the header of the modified table
display(df.head())
# to check, let's count the number of rows for each income category
display(df['floor_category'].value_counts())
```

Calculate and add to the table the ratio of residential and total area, as well as the ratio of kitchen area to total.

```python
# Calculate and add the ratio to the table 
# living_area living space 
# total_area total area
df['living_total_area'] = df['living_area'] / df['total_area']
display(df['living_total_area'].head(3))
```

```python
# Calculate and add the ratio to the table 
# kitchen_area kitchen squares
# total_area total area
df['kitchen_total_area'] = df['kitchen_area'] / df['total_area']
display(df[['living_total_area', 'kitchen_total_area']].head(3))
```

### remove outliers and abnormal data

What features were discovered?

To search for abnormal values, we will create a function that will output:
- Average value
- Median value
- Minimum value
- Schedule

Information about the data:
- first_day_exposition — date of publication
- floor — floor
- floors_total — total floors in the house
- is_apartment — apartments (boolean type)
- kitchen_area — kitchen area in square meters (m2)

```python
# consider all the data data
# to do this, create a function and calculate for each column
# minimum value, maximum value, median value average value
# it is logical to assume that based on the average value, it will be possible 
# to immediately find columns with abnormally large value
columns_for_analysis = ['total_images', 'last_price', 'total_area', 'first_day_exposition',
       'rooms', 'ceiling_height', 'floors_total', 'living_area', 'floor',
       'studio', 'open_plan', 'kitchen_area', 'balcony',
       'locality_name', 'airports_nearest', 'city_сenter_nearest',
       'parks_around3000', 'parks_nearest', 'ponds_around3000',
       'ponds_nearest', 'days_exposition']

count_column_for_analysis = 0            # count of measurements
wrong_column_for_analysis = 0            # count incorrect data
wrong_list_for_analysis = []             # list for incorrect data 

for index in columns_for_analysis:
    try:
        # print
        print('Average value', index, df[index].mean())
        print('Median value', index, df[index].median())
        print('Minimum value', index, df[index].min())
        print('Maximum value', index, df[index].max())
        print()
        #df[index].hist()                # histogram
        #plt.show()                      # separator
        count_column_for_analysis += 1   # counter
    except:
        wrong_list_for_analysis.append(index)
        wrong_column_for_analysis +=1 

print('Count of measurements', count_column_for_analysis)
print('Incorrect lines', wrong_column_for_analysis)
print('Count of incorrect lines', wrong_column_for_analysis)
print('Incorrect lines', wrong_list_for_analysis)
```
